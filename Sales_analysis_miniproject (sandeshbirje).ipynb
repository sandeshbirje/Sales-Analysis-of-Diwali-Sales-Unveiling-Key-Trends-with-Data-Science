# Exploratory Data Analysis

```python
df.columns
```

### 1. Gender

```python
gender_an=sns.countplot(x="Gender",data=df) # here we pass the parameters inside the countplot x= gender & data=datafreame we we cleaned
# inorder to get the count on the bar we do below operation
for bar in gender_an.containers:
    gender_an.bar_label(bar)
```

```python
#here we want to know how much amout of sales is done on gender basis
sales_gen=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
# to get the amout based gender sales we grouped by the gender and the will do te sum and sort that baseed on the amount in decending order

amn=sns.barplot(x="Gender",y="Amount",data=sales_gen)
for bar in amn.containers:
    amn.bar_label(bar)
```

#### from above graph we can visualize that most of the buyers are females and even the purchesing power of the females are greter than males

### 2. Age

```python
age_an=sns.countplot(data=df,x="Age Group",hue="Gender") # here we pass the parameters inside the countplot x= age group, hue=gender & data=datafreame we we cleaned
# inorder to get the count on the bar we do below operation
for bar in age_an.containers:
    age_an.bar_label(bar)
```

```python
# Total amount vs Age Group

#here we want to know how much amout of sales is done on Age basis
sales_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)

amn=sns.barplot(x="Age Group",y="Amount",data=sales_age)
```

#### from above graph we can visualze that most of the buyers are from age 26-35 years females

### 3. State

```python
# here we see top 10 states having  more orders

order_state=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
# grouped the states based on the order sum

sns.set(rc={"figure.figsize":(15,5)}) 

orders=sns.barplot(x="State",y="Orders",data=order_state)
```

```python
# here we can see top 10 states of  more amount spended

order_state=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
# grouped the states based on the order sum

sns.set(rc={"figure.figsize":(16,5)}) 

orders=sns.barplot(x="State",y="Amount",data=order_state)
```

### from above graph we can see that most of the sales/amount from uttar pradesh ,maharastra,karnataka respectivly

### 4. Marital_Status

```python
MS_an=sns.countplot(data=df,x="Marital_Status") # 1-MEANS UNMARRIED AND 0-MARRIED
sns.set(rc={"figure.figsize":(5,3)})
for bar in MS_an.containers:
    MS_an.bar_label(bar)
```

```python
amn_st=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
# grouped the states based on the order sum

sns.set(rc={"figure.figsize":(6,5)}) 

orders=sns.barplot(x="Marital_Status",y="Amount",data=amn_st,hue="Gender") #  1-MEANS UNMARRIED AND 0-MARRIED
```

### from above graph we can see most of the buyes are married(females) and they have high purches power

### 5. Occupation

```python
OC_an=sns.countplot(data=df,x="Occupation") 
sns.set(rc={"figure.figsize":(23,6)})
for bar in OC_an.containers:
    OC_an.bar_label(bar)
```

```python
# occupation vs amount

OC_amn=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)

sns.set(rc={"figure.figsize":(20,5)}) 

orders=sns.barplot(x="Occupation",y="Amount",data=OC_amn)
```

### from above graph we can see that most of the buyers from IT sector,Healthcare, Aviation

### 6. Product Category

```python
# product categery
pd_an=sns.countplot(data=df,x="Product_Category") 
sns.set(rc={"figure.figsize":(25,8)})
for bar in pd_an.containers:
    pd_an.bar_label(bar)
```

```python
# product vs amount

pd_amn=df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(15)

sns.set(rc={"figure.figsize":(28,10)}) 

orders=sns.barplot(x="Product_Category",y="Amount",data=pd_amn)
```

### from above graph we can see that most of the products are from feood, cloathing and electonics but the more amount is spend on food

```python
# product_id vs  orders

pd_ord=df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(15)

sns.set(rc={"figure.figsize":(20,6)}) 

orders=sns.barplot(x="Product_ID",y="Orders",data=pd_ord)
```
